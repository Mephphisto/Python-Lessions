Der "for" loop:

```
print("Hi")
print("Hi")
print("Hi")
```

Wie du dir schon denken kannst das nicht besonders sinvoll!

```
for k in range(3):
  print("Hi")
```

das sieht doch schon besser aus

Beim programmieren erlaubt uns ein loop in kompakter weise viele repititive schritte auf einmal zu programmieren.

python folgt hier dem "for all" schema.

> for [index variable] in [liste]:
>
>>tu etwas mit der [index variablen]

hier führen zuerst eine [index variable] ein, die stellvertretend für alle einträge einer liste steht. und dann können wir eine operation mit der [index variable] programmieren, die von programm für alle einträge durchgeführt wird.

Im obrigen beispiel, deffinieren wir zuerst die [index variable]->k und legen fest das sie für alle einträge in range(10) stehen soll. Für den Moment können wir annehemn, dass range(10) einfach für die Liste alle positiven zahlen kleiner als 10 steht. Später mehr zu range(...).
Im nächsten schritt deffiniern wir, dass wir alle zahlen ausgeben möchten.

```
for k in range(10):
  print("k = " + str(k))
```

Du kannst auch mehrere Pools verschachteln

```
for l in range(3):
  for m in range(2):
    print("l = " + str(l) + " , m = " + str(m))
```
